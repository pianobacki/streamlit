### Task3: Plotting
1.  Display a scatterplot: bill_length_mm vs bill_depth_mm by species

```python
fig, ax = plt.subplots()
sns.scatterplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    ax=ax
)
plt.show()
```

### Task4
1. Determine the average of the bill_length_mm by species

```python
bill_length_mean = df.groupby(by=['species'])['bill_length_mm'].mean()
print(bill_length_mean)
```
